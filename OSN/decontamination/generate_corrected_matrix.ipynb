# 1. generate correted matrix in .h5ad format
Different tools have different formats of output corrected matrix.

rawdata: use the annotated .rds

scCDC: it output a dataframe, we save as .rds by 'saveRDS'

decontx: a matrix in .rds format

soupx: a matrix in .rds format

cellbender: .h5 

fastcar: originally output the matrix, we processed it into seurat object and convert to .h5ad by 'sceasy::convertFormat(fastcar_seurat_obj, from = "seurat", to = "anndata", outFile = "path/FastCAR/MOE_corrected_mat.h5ad")'

scar: it generate 'filtered_feature_bc_matrix_denoised_mRNA.h5ad' directly

cellclear: it output a folder including barcodes.tsv.gz, features.tsv.gz and matrix.tsv.gz, we use 'adata = sc.read_10x_mtx(path, var_names="gene_symbols", cache=True)' to generate matrix



## generate data from .rds

In the process, we met the problem that seurat version is not match. This part fits rawdata, scCDC. The rds is seurat object. 


```python
library(Seurat)
seurat_obj <- readRDS("path/xx.rds")

# if .rds is seurat version 5, but env is seurat version 4

# 1. withdraw the count matrix from seurat obj
counts_mat <- GetAssayData(seurat_obj, assay = "RNA", layer = "counts") # v5 style
# 2. create new seurat object only with count(v4)
seurat_obj <- CreateSeuratObject(counts_mat)
# 3. convert
sceasy::convertFormat(seurat_obj, from = "seurat", to = "anndata", outFile = "path/xx.h5ad", main_layer = "counts")

# if output .rds and env are all v4, convert directly
sceasy::convertFormat(seurat_obj, from = "seurat", to = "anndata", outFile = "path/xx.h5ad")
```
